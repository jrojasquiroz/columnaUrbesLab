```python
import pandas as pd
import geopandas as gpd
import os

os.chdir("C:/Users/rojas/Desktop/Chimbote")

gdf=gpd.read_file("data/1-Mzs_Chimbote_Oficial_INEI.geojson")
refri=pd.read_csv("data/2-Refrigeradora.csv")
hacin=pd.read_csv("data/3-Hacinamiento.csv")
```

```python
refri.rename(columns={'Cod':'Mz','No_TotalPR':'NoRefri_TotalPR'},inplace=True)
hacin.rename(columns={'Cod':'Mz'},inplace=True)
```

```python
gdf_nuevo = gpd.GeoDataFrame(gdf.merge(refri, on='Mz', how='left'))
gdf_nuevo = gpd.GeoDataFrame(gdf_nuevo.merge(hacin, on='Mz', how='left'))
```
